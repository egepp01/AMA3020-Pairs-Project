```python
import numpy as np
import matplotlib.pyplot as plt

m = 9.109 * 10**(-31)
e = 1.602 * 10**(-19)
h = 1.054 * 10**(-34)
n = np.linspace(0.7,10,94)
E = -(m * e**4)/(2 * h**2 * n**2)
Eval = np.zeros((2,10))
for k in range(10):
  Eval[:,k] = -(m * e**4)/(2 * h**2 * (k+1)**2)
plt.plot(n,E); plt.ylabel('Energy'); plt.xlabel('r')
for k in range(10):
  plt.plot([0.7,k+1],Eval[:,k],'g')
plt.legend(); plt.show()
```
